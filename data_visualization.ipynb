### **Univariate Plots**

1. **Histogram**:  
   - **Purpose**: To visualize the frequency distribution of a single numerical variable.  
   - **Example**: Distribution of runs scored across matches.  

2. **Box Plot**:  
   - **Purpose**: To summarize the distribution of a numerical variable, highlight medians, quartiles, and outliers.  
   - **Example**: Spread of strike rates or averages.

3. **Bar Plot**:  
   - **Purpose**: To compare the count or value of categories in a single categorical variable.  
   - **Example**: Runs scored when the toss was won vs. lost.  

4. **Pie Chart**:  
   - **Purpose**: To show proportions of a whole for categorical variables.  
   - **Example**: Proportion of matches played with and without a toss win.

---

### **Bivariate Plots**

1. **Scatter Plot**:  
   - **Purpose**: To visualize the relationship between two numerical variables.  
   - **Example**: Runs scored vs. batting strike rate.

2. **Line Plot**:  
   - **Purpose**: To analyze trends over time or sequences.  
   - **Example**: Runs scored across innings or matches.  

3. **Heatmap**:  
   - **Purpose**: To show correlations between multiple numerical variables.  
   - **Example**: Correlation between runs, boundaries, strike rate, and averages.  

4. **Grouped Bar Plot**:  
   - **Purpose**: To compare multiple numerical variables grouped by a categorical variable.  
   - **Example**: Number of centuries and fifties for toss outcomes.  

5. **Stacked Bar Chart**:  
   - **Purpose**: To show part-to-whole relationships for multiple numerical variables.  
   - **Example**: Contribution of fours and sixes to total boundaries.

```python
df = df.replace("-", 0)
```

```python
data = pd.DataFrame({
    'Info': df["Info"],
})
```

```python
# Create the 'Not_Out' column: 1 if '*' is in 'HS', else 0
data['HS_Not_Out'] = df['HS'].apply(lambda x: 1 if '*' in str(x) else 0)

# Optionally, clean the 'HS' column to remove '*'
data['HS'] = df['HS'].apply(lambda x: str(x).replace('*', '').strip())
```

```python
data["4s"]= df["4s"].astype(int)
data["6s"]=df["6s"].astype(int)
data["0s"]=df["0s"].astype(int)
data["50s"]=df["50s"].astype(int)
data["100s"]=df["100s"].astype(int)
data["SR"]=df["SR"].astype(float)
data["BF"]=df["BF"].astype(int)
data["Avg"]=df["Avg"].astype(float)
data["Runs"]=df["Runs"].astype(int)
data["NO"]=df["NO"].astype(int)
data["Inns"]=df["Inns"].astype(int)
data
```

```python
df.info()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame

# Filter rows where 'Info' contains 'vs'
filtered_by_opp = data[data["Info"].str.contains("vs")]

# Extract the columns needed
teams = filtered_by_opp["Info"]
innings = filtered_by_opp["Inns"]

# Plot the bar chart
x = range(len(teams))  # Positions for the bars

plt.figure(figsize=(10, 6))
plt.bar(x, innings, width=0.4, label="Innings", color="skyblue", align="center")

# Add labels and title
plt.title("Innings vs Centuries by Virat Kohli Against Teams", fontsize=14)
plt.xlabel("Opposition Teams", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(x, teams, rotation=45, ha="right")
plt.legend()

# Add gridlines
plt.grid(axis="y", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

## Centuries per year

```python
import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')

import pandas as pd
import matplotlib.pyplot as plt

# Filter rows where 'Info' contains 'year'
filtered_by_year = data[data["Info"].str.contains("year")]

# Extract the year from the 'Info' column
filtered_by_year.loc[:, "Year"] = filtered_by_year["Info"].str.extract(r'(\d{4})').astype(int)


# Sort by year to ensure chronological order
filtered_by_year = filtered_by_year.sort_values(by="Year")

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(filtered_by_year["Year"], filtered_by_year["100s"], marker="o", linestyle="-", color="orange", label="Centuries")

# Add labels and title
plt.title("Centuries Per Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Centuries", fontsize=12)
plt.xticks(filtered_by_year["Year"], rotation=45)
plt.legend()

# Add gridlines
plt.grid(axis="both", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

1. The chart shows the number of centuries scored per year, with peaks in 2011, 2016, and 2018, reaching up to 6 centuries in those years.  
2. A significant dip is observed in 2014, 2020, and 2024, where no centuries were scored.  
3. The overall trend reflects fluctuating performance, with stable contributions in 2012–2013 and sharp improvements in peak years.  

### Runs scored per team

```python

import warnings

# Suppress all warnings
warnings.filterwarnings('ignore')


# Filter rows where 'Info' contains 'year'
filered_by_year = data[data["Info"].str.contains("year")]

# Extract the year from the 'Info' column
filered_by_year["Year"] = filered_by_year["Info"].str.extract(r'(\d{4})').astype(int)

# Sort by year to ensure chronological order
filered_by_year = filered_by_year.sort_values(by="Year")

# Plot the line chart
plt.figure(figsize=(10, 6))
plt.plot(filered_by_year["Year"], filered_by_year["Runs"], marker="o", linestyle="-", color="orange", label="Runs")

# Add labels and title
plt.title("Runs Per Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Runs", fontsize=12)
plt.xticks(filered_by_year["Year"], rotation=45)
plt.legend()

# Add gridlines
plt.grid(axis="both", linestyle="--", alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```

1. This chart illustrates the total runs scored per year highlighting peaks in 2011, 2016, and 2018 with over 1,400 runs, 
2. indicating exceptional performance during these years. Significant dips in 2008, 2020, and 2024 suggest fewer matches or external factors like disruptions.
3.  The overall trend shows fluctuating consistency, with stable performances between 2013–2015 and 2017–2019.

```python
# Grouped Bar Plot
df_grouped = filtered_by_year.groupby("Year")[["Runs", "BF"]].sum().reset_index()

# Plot grouped bar chart
df_grouped.plot(x="Year", kind="bar", figsize=(10, 6), stacked=False, color=["skyblue", "orange"])
plt.title("Total Runs and Matches by Year", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend(["Runs", "BF"])
plt.show()
```

```python
# Bubble plot
plt.figure(figsize=(10, 6))
plt.scatter(filtered_by_year["Inns"], filtered_by_year["Runs"], s=filtered_by_year["Avg"]*10, alpha=0.6, c=filtered_by_year["Year"], cmap="viridis")
plt.colorbar(label="Year")
plt.title("Bubble Plot: Matches vs Runs", fontsize=14)
plt.xlabel("Matches", fontsize=12)
plt.ylabel("Runs", fontsize=12)
plt.show()
```

```python
# Line plot for Runs and Matches over Years
plt.figure(figsize=(10, 6))
sns.lineplot(x="Year", y="Runs", data=filtered_by_year, marker="o", label="Runs")
sns.lineplot(x="Year", y="Inns", data=filtered_by_year, marker="o", label="Matches")
plt.title("Runs and Matches Over Years", fontsize=14)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.legend()
plt.grid()
plt.show()
```
